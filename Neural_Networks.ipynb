````python
##### 4 #####

## Build the model
##

import torch.nn as nn

class Model(nn.Module): ## define the number of features input and output
  def __init__(self):
    super(Model,self).__init__()
    self.fc1 =torch.nn.Linear(x.shape[1],5) # x.shape[1] to get the number of columns (features) // convert them to 5 outputs
    self.fc2 =torch.nn.Linear(5,3) # then receive the 5 outputs into 3
    self.fc3 =torch.nn.Linear(3,2) # from 3 into 2 (2 because we need only 2 labels "0,1")
    self.sigmoid=torch.nn.Sigmoid()

  def forward(self,x):
    #
    out =self.fc1(x)
    out =self.sigmoid(out) # we use sigmoid here for the activation function, you can replace it with other liner or non-linear functions
    out =self.fc2(out)
    out =self.sigmoid(out)
    out =self.fc3(out)
    return out
````

````python
For binary classification you should can use the BCELoss or BCEWithLogitsLoss. The difference between the two is that BCELoss assumes the input is probabilities that are computed using nn.Sigmoid() in your last layer. On the other hand, BCEWithLogitsLoss assumes that the inputs are logits without passing through Sigmoid activation.

Finally, in both cases (BCELoss or BCEWithLogitsLoss) the input and target should be 1-dimensional tensors with size [batch_size].
````

````python
import torch.nn as nn
class CNN(nn.Module):
  def __init__(self):
    super(CNN,self).__init__()
    # input_size: 28
    self.cnn1=nn.Conv2d(in_channels=1, out_channels=8, kernel_size=3, stride=1, padding=1)
    # input_size-filter_size +2(padding)/stride + 1 = 28-3+2(1)/1+1=28

    self.batchnorm1=nn.BatchNorm2d(8)
    # output_channel:8, batch(8)

    self.relu=nn.ReLU()
    self.maxpool1=nn.MaxPool2d(kernel_size=2)
    #input_size=28/2=14

    self.cnn2=nn.Conv2d(in_channels=8, out_channels=32, kernel_size=5, stride=1, padding=2)
    # same_padding: (5-1)/2=2:padding_size.
    self.batchnorm2=nn.BatchNorm2d(32)
    self.maxpool2=nn.MaxPool2d(kernel_size=2)
    # input_size=14/2=7
    # 32x7x7=1568

    self.fc1 =nn.Linear(in_features=1568, out_features=600)
    self.fc2 =nn.Linear(in_features=600, out_features=10)
  def forward(self,x):
    print('input shape', x.shape)
    out =self.cnn1(x)
    print('Cnn1 output', out.shape) # N, C, H, W
    out =self.batchnorm1(out)
    print('batch', out.shape)
    out =self.relu(out)
    out =self.maxpool1(out)
    print('max pool shape', out.shape) # 100, 8, 14,14
    out =self.cnn2(out)
    print('cnn2 shape', out.shape) # 100, 32, 14,14
    out =self.batchnorm2(out)
    out =self.relu(out)
    out =self.maxpool2(out)
    print('max pool 2',out.shape) # 100, 32, 7, 7
    out =out.view(-1,1568)
    out =self.fc1(out)
    out =self.relu(out)
    out =self.fc2(out)
    return out
````

````python
class ConvNet(nn.Module):
    def __init__(self, kernels, classes=10):
        super(ConvNet, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(1, kernels[0], kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(8),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2))
        self.layer2 = nn.Sequential(
            nn.Conv2d(8, kernels[1], kernel_size=5, stride=1, padding=2),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2))
        self.fc = nn.Linear(7 * 7 * kernels[-1], classes)

    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = out.reshape(out.size(0), -1)
        out = self.fc(out)
        return out
````

````python
def model_pipeline(hyperparameters):

    # tell wandb to get started
    with wandb.init(project="pytorch-demo", config=hyperparameters):
      # access all HPs through wandb.config, so logging matches execution!
      config = wandb.config

      # make the model, data, and optimization problem
      model, train_loader, test_loader, criterion, optimizer = make(config)
      print(model)

      # and use them to train the model
      train(model, train_loader, criterion, optimizer, config)

      # and test its final performance
      test(model, test_loader)

    return model
````

````python
def make(config):
    # Make the data
    train, test = get_data(train=True), get_data(train=False)
    train_loader = make_loader(train, batch_size=config.batch_size)
    test_loader = make_loader(test, batch_size=config.batch_size)

    # Make the model
    model = ConvNet(config.kernels, config.classes).to(device)

    # Make the loss and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer=torch.optim.SGD(model.parameters(), lr=config.learning_rate)

    return model, train_loader, test_loader, criterion, optimizer
````

````python
from tqdm import tqdm
def get_data(slice=5, train=True):
    full_dataset= datasets.MNIST(root='./data', train=True, transform=transforms.ToTensor(), download=True)
    #  equiv to slicing with [::slice]
    sub_dataset = torch.utils.data.Subset(
      full_dataset, indices=range(0, len(full_dataset), slice))

    return full_dataset


def make_loader(dataset, batch_size):
    loader=torch.utils.data.DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
    return loader


def train(model, loader, criterion, optimizer, config):
    # tell wandb to watch what the model gets up to: gradients, weights, and more!
    wandb.watch(model, criterion, log="all", log_freq=10)

    # Run training and track with wandb
    total_batches = len(loader) * config.epochs
    example_ct = 0  # number of examples seen
    batch_ct = 0
    for epoch in tqdm(range(config.epochs)):
        for _, (images, labels) in enumerate(loader):

            loss = train_batch(images, labels, model, optimizer, criterion)
            example_ct +=  len(images)
            batch_ct += 1

            # Report metrics every 25th batch
            if ((batch_ct + 1) % 25) == 0:
                train_log(loss, example_ct, epoch)


def train_batch(images, labels, model, optimizer, criterion):
    images, labels = images.to(device), labels.to(device)

    # Forward pass ➡
    outputs = model(images)
    loss = criterion(outputs, labels)

    # Backward pass ⬅
    optimizer.zero_grad()
    loss.backward()

    # Step with optimizer
    optimizer.step()

    return loss

def train_log(loss, example_ct, epoch):
    loss = float(loss)

    # where the magic happens
    wandb.log({"epoch": epoch, "loss": loss}, step=example_ct)
    print(f"Loss after " + str(example_ct).zfill(5) + f" examples: {loss:.3f}")

def test(model, test_loader):
    model.eval()

    # Run the model on some test examples
    with torch.no_grad():
        correct, total = 0, 0
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        print(f"Accuracy of the model on the {total} " +
              f"test images: {100 * correct / total}%")

        wandb.log({"test_accuracy": correct / total})

    # Save the model in the exchangeable ONNX format
    torch.onnx.export(model, images, "model.onnx")
    wandb.save("model.onnx")
````

# **Feed Forward Neural Networks**

- It has uni-directional forward propagation but no backward propagation. Weights are static here. An activation function is fed by inputs which are multiplied by weights.

````python
# call torch and Neural Networks NN
import torch
import torch.nn as nn
import torch.nn.functional as F

# call optimizer
from torch.optim import SGD # SGD => stochastic gradient discent

# call graph
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Create function neural network
class FirstNN(nn.Module): # nn.Module to convert the ordinary python class into NN

    def __init__(self):

        super().__init__() # "super" refere to nn.Module above

        # define the place holders (neural sections)
        self.w00 = nn.Parameter( torch.tensor(1.1), requires_grad = False ) # convert the value into Tensor (self.w00 = 1.1) //  requires_grad = False => mean that 1.1 is constant value & not need to decient gradient
        self.b00 = nn.Parameter( torch.tensor(0.6), requires_grad = False )

        self.w01 = nn.Parameter(torch.tensor(0.6), requires_grad=False)
        self.b01 = nn.Parameter(torch.tensor(-1.5), requires_grad=False)

        self.w02 = nn.Parameter(torch.tensor(0.5), requires_grad=False)
        self.b02 = nn.Parameter(torch.tensor(0.65), requires_grad=False)

        self.w11 = nn.Parameter(torch.tensor(1.55), requires_grad=False)
        self.w12 = nn.Parameter(torch.tensor(-0.7), requires_grad=False)
        self.w13 = nn.Parameter(torch.tensor(1.22), requires_grad=False)

        self.w20 = nn.Parameter(torch.tensor(1.6), requires_grad=False)

        # forward => draw the function that will apply on the neural
    def forward(self, input):

          input_to_layer_01 = input * self.w00 + self.b00
          layer_01_output = F.relu(input_to_layer_01)  # call the activation function
          layer_01_output = layer_01_output * self.w11

          input_to_layer_02 = input * self.w01 + self.b01
          layer_02_output = F.relu(input_to_layer_02)
          layer_02_output = layer_02_output * self.w12

          input_to_layer_03 = input * self.w02 + self.b02
          layer_03_output = F.relu(input_to_layer_03)
          layer_03_output = layer_03_output * self.w13

          output = layer_01_output + layer_02_output + layer_03_output

          output = F.tanh(output) # Function of activation function
          output = output * self.w20

          return output
````

# **Multilayer Perceptron**

````python
# Let's train and add gredient decient
# Create function neural network
class PlainFirstNN(nn.Module): # nn.Module to convert the ordinary python class into NN

    def __init__(self):

        super().__init__() # "super" refere to nn.Module above

        # define the place holders (neural sections)
        self.w00 = nn.Parameter( torch.tensor(0.0), requires_grad = True ) # requires_grad = True => mean that we need decient gradient to predict the best values
        self.b00 = nn.Parameter( torch.tensor(0.6), requires_grad = False )

        self.w01 = nn.Parameter(torch.tensor(0.6), requires_grad=False)
        self.b01 = nn.Parameter(torch.tensor(-1.5), requires_grad=False)

        self.w02 = nn.Parameter(torch.tensor(0.0), requires_grad=True)
        self.b02 = nn.Parameter(torch.tensor(0.65), requires_grad=False)

        self.w11 = nn.Parameter(torch.tensor(1.55), requires_grad=False)
        self.w12 = nn.Parameter(torch.tensor(-0.7), requires_grad=False)
        self.w13 = nn.Parameter(torch.tensor(1.22), requires_grad=False)

        self.w20 = nn.Parameter(torch.tensor(0.0), requires_grad=True)

        # forward => draw the function that will apply on the neural
    def forward(self, input):

          input_to_layer_01 = input * self.w00 + self.b00
          layer_01_output = F.relu(input_to_layer_01)  # call the activation function
          layer_01_output = layer_01_output * self.w11

          input_to_layer_02 = input * self.w01 + self.b01
          layer_02_output = F.relu(input_to_layer_02)
          layer_02_output = layer_02_output * self.w12

          input_to_layer_03 = input * self.w02 + self.b02
          layer_03_output = F.relu(input_to_layer_03)
          layer_03_output = layer_03_output * self.w13

          output = layer_01_output + layer_02_output + layer_03_output

          output = F.tanh(output) # Function of activation function
          output = output * self.w20

          return output
````

````python
plain_first_nn = PlainFirstNN()
````

**We need to reach to correct graph by calculting the loss between true_output & pred_output till reach to the correct weights**

````python
# Firstly we need to define the optimizer to help the model in start producing the right values
optimizer = SGD(plain_first_nn.parameters(), lr = 0.01) # lr = learning rate
loss = nn.MSELoss() # calculate the loss by taking the mean square error
````

````python
# detect the number of epochs "= number of iteration on loss function to get the right values"
for epoch in range(100):

  total_loss = 0

  for i in range(len(true_inputs)):

    input_i = true_inputs[i]
    true_outputs_i = true_outputs[i]

    pred_output_i = plain_first_nn(input_i)

    loss_value = loss(pred_output_i, true_outputs_i)

    loss_value.backward() # to edit the value that needed gradient decent

    total_loss += loss_value

  print("Epoch:", epoch," | Total Loss:",total_loss)

  # after finish this epoch => make optimizer to take another step in learning
  optimizer.step()
  # we need to renew the calcultion of optimzer to take step from zero in every iterate
  optimizer.zero_grad()
````
