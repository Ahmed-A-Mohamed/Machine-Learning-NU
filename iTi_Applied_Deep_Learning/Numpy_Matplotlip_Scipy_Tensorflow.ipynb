# **Matplotlip**

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.title('sine and cosine curves')
plt.legend(['Sine','Cosine'])
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

```python
# Subplotting
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1) # hight = 1 & width = 1 & the number of active plot = 1
plt.plot(x,y_sin)
plt.title('Sine curve')

plt.subplot(2,1,2) # hight = 1 & width = 1 & the number of active plot = 2
plt.plot(x,y_cos)
plt.title('Cosine curve')


plt.show()
```

```python
import tensorflow as tf
```

```python
# Run the session of Tensorflow (TensorFlow is not coding interface but like terminal session)
import tensorflow as tf

tf.compat.v1.disable_eager_execution()

a = 2
b = 3
c = tf.add(a, b, name='add')

with tf.compat.v1.Session() as sess:
    result = sess.run(c)
    print(result)
```

`tf.compat.v1.disable_eager_execution()`  Use the session-based approach


```python
import tensorflow as tf

tf.compat.v1.disable_eager_execution()  # Use the session-based approach

x = 2
y = 3
add_op = tf.add(x, y, name='add')  # add x + y & name the operation "add"
mult_op = tf.multiply(x, y, name='multiply')
pow_op = tf.pow(add_op, mult_op, name='power')
res = tf.multiply(x, pow_op, name="result")

with tf.compat.v1.Session() as sess:
    # Run all necessary operations
    pout = sess.run([add_op, mult_op, pow_op, res])

# Print the results
print("Addition:", pout[0])
print("Multiplication:", pout[1])
print("Power:", pout[2])
print("Result:", pout[3])
```

choosing `name= ` of operation depend on spicific name. If you don't know the exact names use `name=Useless`
